```python
import matplotlib.pyplot as plt
```

```python
labels = ["sachin", "shewag", "virat", "Yuvraj"]
runs = [10000, 4000, 7000, 11000]
```

```python
plt.title("career Runs of sachin, shewag, virat, yuvraj")
plt.pie(runs, labels=labels)
plt.show()
```

# Custom colors

```python
colors = ["#2bfb47", "#f91212", "#a502fb", "#f9ee12"]
plt.pie(runs, colors=colors, labels=labels,  wedgeprops = {"edgecolor":"Blue", "linewidth":2,  "linestyle":'--'})
```

```python
explode = [0., 0,0,0]
plt.pie(runs, colors=colors, labels=labels, explode=explode, shadow=True)
```
