# Lets compair sachin, shewag, and kohli side by side in the same year

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
years = [2000,2001,2002,2003,2004,2005]

sachin = [1000,15000,25000,30000,45000,50000]
shewag = [2500,8500,14000,15000,20000,23000]
virat = [5000,8000,11000,16000,25000,31000]
```

```python
x  = np.arange(len(years))
width = 0.25
```

```python
plt.bar(x-width, sachin, width=width, label="sachin")
plt.bar(x, shewag, width, label="Shewag")
plt.bar(x+width, virat, width=width, label="virat")
plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Run comparison betweem sachin, shewag, and virat")
# to put years as xlabel then we use 
plt.xticks(x, years)
# To make ato adjust space then we use of 
plt.tight_layout()

plt.legend()
plt.show
```
