```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.info()
```

# Line Plot

```python
tips = sns.load_dataset('tips')
sns.lineplot(x="total_bill", y="tip", data=tips)
plt.title("Line plot Examples")
plt.show()
```

# Scatter Plot

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="time")
```

# Bar Plot

```python
tips.info()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips, color="red")
```

# Box Plot

```python
tips.info()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
```

# Pivot Table

```python
print(sns.get_dataset_names())
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
pv = flights.pivot(index="month", columns="year", values="passengers")
```

```python
pv
```

# Heatmap

```python
sns.heatmap(pv, annot=True,fmt="d")
plt.title("Heatmap display")
```
